## Analyze traffic accident data to identify patterns related to road conditions, weather, and time of day. Visualize accident hotspots and contributing factors.

### 

```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
data=pd.read_csv('traffic_accidents_data.csv')
```

```python
# Step 1: Handling Missing Values
data.isnull().sum()
     
```

```python
# Step 2: Handling Duplicates
# Drop duplicate rows
data.drop_duplicates(inplace=True)
    
```

```python
# Convert 'Date' and 'Time' columns to datetime format
data['DateTime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'])

# Set 'DateTime' as the index
data.set_index('DateTime', inplace=True)
```

```python
# Accident Hotspots (Location)
sns.scatterplot(x='Longitude', y='Latitude', data=data, alpha=0.1)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
     
```

```python
# Road Conditions vs. Accident Count
plt.figure(figsize=(10, 6))
sns.countplot(x='Road Conditions', data=data)
plt.title('Accidents by Road Conditions')
plt.xlabel('Road Conditions')
plt.ylabel('Accident Count')
plt.show()
```

```python
# Weather vs. Accident Count
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather Conditions', data=data)
plt.title('Accidents by Weather')
plt.xlabel('Weather')
plt.ylabel('Accident Count')
plt.show()
     
```

```python
# Time of Day vs. Accident Count
plt.figure(figsize=(10, 6))
sns.countplot(x='Time', data=data)
plt.title('Accidents by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Accident Count')
plt.show()
```

```python
# Distribution of accidents based on month
plt.figure(figsize=(10, 6))
data.index.month.value_counts().sort_index().plot(kind='bar',color="cyan")
plt.title('Distribution of Accidents based on Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.xticks(range(1, 13), ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data.index.dayofweek.value_counts().sort_index().plot(kind='bar',color = "yellow")
plt.title('Distribution of Accidents based on Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], rotation=45)
plt.show()
```
